```python
years = [1999,2000,2001,2002,2003]
happy = [.8,.9,.3,.4,.6]
sad = [.2,.1,.7,.6,.4]
```

```python
plt.bar(years,happy)
plt.show()
```

```python
plt.plot(years,happy)
plt.plot(years,sad)
plt.legend(("Happy","Sad"))
plt.show()
```
